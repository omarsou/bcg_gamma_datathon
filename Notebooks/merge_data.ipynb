```python
import pandas as pd
import numpy as np
```

# Creating the big dataframe

## Load data

```python
df_meteo = pd.read_csv('../Data/meteo_absolute_final.csv', sep=';').reset_index(drop=True)
df_vacances = pd.read_csv('../Data/data_vacances_ferie.csv', sep=';', encoding = 'ISO-8859-1').reset_index(drop=True)
df_covid = pd.read_csv('../Data/data_covid.csv', sep=';').reset_index(drop=True)
df_greves = pd.read_csv('../Data/greves.csv', sep=';').drop('Unnamed: 0', axis=1).reset_index(drop=True)
df_champs = pd.read_csv('../Data/ChampsElyseesComplete/champs_complet.csv').drop('Unnamed: 0', axis=1).reset_index(drop=True)
df_convention = pd.read_csv('../Data/ConventionComplete/convention_complet.csv').drop('Unnamed: 0', axis=1).reset_index(drop=True)
df_saintp = pd.read_csv('../Data/SaintsPeresComplete/saintsperes_complet.csv').drop('Unnamed: 0', axis=1).reset_index(drop=True)
```

```python
df_champs_last = pd.read_csv('../Data/ChampsElyseesComplete/champs_last.csv').drop('Unnamed: 0', axis=1)
df_convention_last = pd.read_csv('../Data/ConventionComplete/convention_last.csv').drop('Unnamed: 0', axis=1)
df_saintp_last = pd.read_csv('../Data/SaintsPeresComplete/saintsperes_last.csv').drop('Unnamed: 0', axis=1)
```

```python
df_champs_last = df_champs_last[df_champs.columns.tolist()]
df_convention_last = df_convention_last[df_convention.columns.tolist()]
df_saintp_last = df_saintp_last[df_saintp.columns.tolist()]
```

```python
df_champs = pd.concat([df_champs, df_champs_last]).reset_index(drop=True)
df_convention = pd.concat([df_convention, df_convention_last]).reset_index(drop=True)
df_saintp = pd.concat([df_saintp, df_saintp_last]).reset_index(drop=True)
```

## Refacto Data (and some simple features engineering)

## Greves

```python
df_greves.intensite_greve = df_greves.intensite_greve.fillna(np.float64(0))
df_greves['t_1h'] = pd.to_datetime(df_greves['t_1h']) ## Convert date to datetime
df_greves.head()
```

```python
df_greves.columns = ['Datetime', 'intensite_greve']
df_greves.tail()
```

### Meteo

```python
df_meteo.info()
```

```python
def replace_comma(x):
    try:
        x = x.replace(',','.')
    except AttributeError:
        return x
    return x
df_meteo.temp = df_meteo.temp.apply(replace_comma) ## REPLACE "," by "."
def convert_to_float64(x):
    return x.astype(np.float64)
df_meteo['Datetime'] = pd.to_datetime(df_meteo['Datetime']) ## Convert date to datetime
## Convert to float64
df_meteo[['temp', 'rain', 'fog', 'thunder', 'snow']] = df_meteo[['temp', 'rain', 'fog', 'thunder', 'snow']].apply(convert_to_float64) 
```

```python
df_meteo.info()
```

### Vacances

```python
df_vacances.info()
```

```python
df_vacances['Datetime'] = pd.to_datetime(df_vacances['Datetime'])
cols_to_convert = [x for x in df_vacances.columns.tolist() if x != 'Datetime']
df_vacances[cols_to_convert] = df_vacances[cols_to_convert].apply(convert_to_float64)
```

### Covid

```python
df_covid.info()
```

```python
df_covid['Datetime'] = pd.to_datetime(df_covid['Datetime'])
cols_to_convert = [x for x in df_covid.columns.tolist() if x != 'Datetime']
df_covid[cols_to_convert] = df_covid[cols_to_convert].apply(convert_to_float64)
```

### Road traffic

```python
df_champs.tail()
```

```python
df_champs.describe()
```

```python
cols_to_keep = ['t_1h', 'etat_barre', 'q', 'k']
df_champs = df_champs[cols_to_keep]
df_convention = df_convention[cols_to_keep]
df_saintp = df_saintp[cols_to_keep]
```

```python
df_champs.columns = ['Datetime', 'etat_barre_ce', 'q_ce', 'k_ce']
df_convention.columns = ['Datetime', 'etat_barre_lc', 'q_lc', 'k_lc']
df_saintp.columns = ['Datetime', 'etat_barre_pv', 'q_pv', 'k_pv']
```

```python
df_champs['Datetime'] = pd.to_datetime(df_champs['Datetime'])
df_convention['Datetime'] = pd.to_datetime(df_convention['Datetime'])
df_saintp['Datetime'] = pd.to_datetime(df_saintp['Datetime'])
```

### Merge all

```python
lst = [df_meteo, df_vacances, df_covid, df_greves, df_champs, df_convention, df_saintp]
big_df = pd.concat([x.set_index("Datetime") for x in lst], axis=1)
```

```python
big_df.to_csv('../Data/all_in_one.csv')
```
